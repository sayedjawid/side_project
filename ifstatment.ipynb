```python
admin = 'lucas'
password = 1234


if admin == 'lucas':
    user_pass = input('Please enter your password:')
    try:
        if int(user_pass) == password:
            print(f'Hi {admin.title()}, you can do everything as you wish!')
        else:
            print('Password is not correct! ')
    except ValueError:
        print('Invalid input! Password must be a number.')
else:
    print('You are not authorized to log in!')

```

```python
users = {
    'lucas': '123456',
    'anna': 'abc123',
    'max': 'qwerty'
}

admin = input("Enter your username: ")

if admin in users:
    user_pass = input("Enter your password: ")
    if user_pass == users[admin]:
        print(f"Hi {admin.title()}, welcome back!")
    else:
        print("Wrong password!")
else:
    print("User not found!")
```

```python
admin = 'lucas'
password = 1234

if admin == 'lucas':
    for attempt in range(3):
        user_pass = input('Please enter your password: ')
        try:
            if int(user_pass) == password:
                print(f'Hi {admin.title()}, you can do everything as you wish!')
                break
            else:
                print('Password is not correct!')
        except ValueError:
            print('Invalid input! Password must be a number.')
    else:
        print('Too many failed attempts. Access denied.')
else:
    print('You are not authorized to log in!')
```
