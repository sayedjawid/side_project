•

• 
Add a line at the end of your program stating what these animals have in common. You could print a sentence such as Any of these animals would make a great pet!

```python
animals = ['cat', 'puma', 'tiger']

for animal in animals:
    print(f'This is a: {animal.title()}')
    if animal == 'cat':
        print("The cat is small and you can have it as a pet.")
    elif animal == 'puma':
        print("The puma is a strong wild cat that lives in the mountains.")
    else:
         print("The tiger is a big striped cat that lives in the jungle.")
    # print(f'The {animal[0]} is small and you can have them at your house.')
```

```python
#  copying list


countries = ['sweden', 'norway', 'finland', 'denmark', 'iceland']
friends_country = countries[:] # create a copy of countries list
countries.append('estonia')

friends_country.append('iran')
print("My countries:", countries)
print("My friend's countries:", friends_country)

countries[0] = 'switzerland'
print("My countries after change:", countries)
print("My friend's countries after my change:", friends_country)

friends_country[0] = 'spain'
print("My countries after friend's change:", countries)
print("My friend's countries after change:", friends_country)
```
